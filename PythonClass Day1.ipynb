1. 숫자형  
숫자형 자료형은 숫자 형태 (정수, 실수 등)로 이루어진 자료형이다.

```python
a = 3
b = 5

# print는 생략해도 출력이 되지만 정상은 print 구문을 사용하는 것이다.
# 주석(comment)은 컴퓨터는 읽지 않는다.
print (a+b) 
print (a-b)
print (a*b)
print (a/b)
```

# **문자열 자료형**

문자열을 문자, 단어, 문장 등으로 구성된 글자들의 집합  


```python
a = "b"
b = "Hello"
c = "awesome class"
d = "123" # 123 숫자 123과는 다른 문자 "123"이다.

print (a)
print (b)
print (c)
print (d)

# 자바 등과 같은 언어에서는 변수를 사용할 때 형까지도 지정을 해주어야만 한다.
# char a = "b";
# String b = "Hello";
# Int c = 5;
```

```python
# 파이썬에서 문자열은 "" 큰따옴표와''작은따옴표는 기본적으로 동일하게 취급한다

a = "Hello"
b = 'Hello'
print (a)
print (b)

# 하지만 아래의 경우 약간 다르게 처리된다
a = 'Hello "Austin"'
print (a)

b = "Hello 'Austin'"
print (b)

# """    """  doc string. 컴퓨터도 읽는 주석
# '''    '''  

# 백슬래시를 넣으면 기능이 무시되고 글자로만 취급된다
a = "Hello \"Austin\""
print (a)
b = 'Hello \'Austin\''
print (b)
```

```python
# \ 백슬래시를 사용하는 것을 이스케이프 코드를 사용한다고 표현한다.
# \'
# \"
# \n 강제 줄바꿈
# \t 일정 간격 띄우기

a = "Hello \nWorld"
print (a)

b = "Hello\t\t\tWorld"
print (b)
```

#LIST [ ]
 - mutable 수정 가능하다.
 - indexing / slicing 가능하다.
 - 파이썬에서 가장 많이 사용된다.

#Tuple ( )
 - immutable 수정 불가능하다.
 - indexing / slicing 가능하다.
 - 요소를 하나만 저장할 때도 쉼표를 붙인다.

#Dic { }
 - mutable 수정 가능하다.
 - indexing / slicing 불가능하다.
 - Key-Value 값 쌍으로 구성된다

#Set { }
 - mutable 수정 가능하다.
 - indexing / slicing 불가능하다.
 - 중복 값을 허용하지 않는다.



```python
# 리스트 작성

emp1 = []
odd2 = [1,3,5,7,9]
str3 = ["하나", "둘","삼"]
mix4 = [1,2,"삼"]

print(emp1)
print(odd2)
print(str3)
print(mix4)

# 리스트에서 인덱싱 indexing (인덱싱은 0에서부터 시작한다)

alist = [10,20,30,40,50]
print(alist[0]) # print를 제거해도 값 출력 된다.
print(alist[0]+alist[1]) #indexing 한 list 값 끼리 계산이 가능하다.
print((alist[0]+alist[1]+alist[2]+alist[3]+alist[4]) / len(alist))  # ()괄호 안에 수식부터 계산하고 다음 나누기 계산
```

```python
# 슬라이싱 scliing 하나식 추출하는 인덱싱과는 달리 여러 개를 추출

blist = "12345"
#blist[0:2] # 0부터 2미만, 2는 포함하지 않는다
#blist[:3]  # 시작을 생략하는 처음부터 3미만까지 3은 포함하지 않는다
blist[0:]  # 마지막을 생략하면 마지막까지
```

```python
# 리스트의 연산

# 리스트 더하기

alist = [1,2,3,4,5]
blist = [6,7,8,9,10]
print (alist+blist)

#리스트 곱하기 (리스트들을 반복하여 연결된 합으로 출력)

print(alist * 3)

#리스트의 갯수 구하기

print(len(blist))
```

```python
# 리스트의 수정 mutable

alist = [1,2,3,4,5]
alist[2] = 0
print (alist)

# 리스트의 삭제

del alist[2]
print(alist)
```

```python
#리스트에서 많이 사용되는 함수들

# append - 리스트에 요소를 추가하는 기능

alist = [1,2,3,4,5]
alist.append(6)
print(alist)
print("원래 리스트에서 요소를 추가한 결과는", alist,"입니다")

# sort - 리스트에서 요소들을 정렬하는 기능

blist = [2,4,3,5,1]
blist.sort()
print (blist)
print("원래 리스트를 정렬한 결과는", blist,"입니다" )

# reverse - 리스트에서 요소들을 그대로 뒤집는 기능

clist = [1,2,3]
clist.reverse()
print(clist)
print("원래 리스트를 그대로 역순으로 출력한 결과는", clist,"입니다" )

# index - 리스트에서 특정 위치의 값을 반환하는 기능

dlist = [1,2,3,4,5]
#print(dlist.index(2)) #2의 값의 위치를 반환하기 때문에 1이 반환 됨, 즉 값을 인식한다는 의미
#print("숫자 2의 인덱스 값은",dlist.index(2),"입니다")

# insert - 리스트의 특정 위치에 요소를 추가하는 함수
elist = [1,2,3,4,5]
elist.insert(0, 0) # (특정 위치 인덱스, 넣을 값)
print(elist)
print("0 위치에 0 값을 넣은 결과는", elist, "입니다")

# remove - 리스트에서 요소를 삭제하는 기능

flist = [1,2,3,4,5,3]
flist.remove(3) #리스트에서 삭제할 값을 입력, 값은 값이 여러 개 있을 경우 처음 등장하는 값을 삭제
print(flist)
print("중복 된 값이 있을 경우 처음 중복 값을 삭제한 결과는", flist, "입니다")

# pop - 리스트에서 마지막 요소를 제거하는 용도로 주로 쓰이지만 특정 위치로도 삭제

glist = [1,2,3,4,5]
glist.pop()
glist.pop()
print(glist)
print("마지막 두 개의 값을 삭제 한 결과는", glist, "입니다")

glist.pop(0) # 0번째 값을 삭제, 위 수식이 진행되었기 때문에 1,2,3이 존재하는 상태에서 0번재 값은 1이다.
print("0번째 값을 삭제한 결과는", glist, "입니다")  

# 리스트에서 요소의 개수 세기 기능 'count'  

jlist = [1,2,3,1]
jlist.count(1)  #리스트 배열에서 입력 값에 대한 개수를 출력
print("1 값의 개수를 출력한 결과는", jlist, "입니다")

# 리스트의 확장은 기존 리스트에 다른 리스트를 더하여 하나의 리스트를 만든다.

a = [1,2,3]
a.extend([4,5])
print(a)
print("기존 리스트에 다른 리스트 값을 더하여 만든 결과는", a, "입니다")
```

```python
# 딕셔너리 Dict

# 이름 그대로 dictionary 사전 처럼 Key:Value 값의 상으로 데이터가 저장되는 자료형이다
# 리스트처럼 mutable 한 속성을 갖고 있으나 인뎅싱과 슬라이싱은 지원하지 않는다.
# 순서가 없고, Key로 값을 Value를 꺼내기 때문에 속도가 빠르다

adict = {5:'a', 3:'b', 2:'c'}

print(adict[5])
print(adict[5], adict[3], adict[2])
print(adict)

adict[9] = 'd' #mutable 한 속성 중 하나로, 값을 추가할 수 있다
print(adict)

del adict[3]   #multable 한 속성 중 하나로 값을 삭제 할 수 있으며, Key 값을 통해 Value를 같이 삭제할 수 있다.
print(adict)
```

```python
adict[0] # Dictionary는 인덱싱과 슬라이싱이 불가능하다
```

```python
adict = {5:'a', 3:'b', 2:'c'}

# 딕셔너리에서 키 값만 추출하는 기능 keys
print(adict.keys())   # Key 값만 추출

# 딕셔너리에서 키 값만 추출하는 기능 keys
print(adict.values()) # Value 값만 추출

# 딕셔너리에서 쌍으로 추출하는 기능 items
print(adict.items())  # 키, 값 쌍으로 추출

# 딕셔너리에 키 값이 존재하는지 확인하는 기능 in
1 in adict
#5 in adict
```
